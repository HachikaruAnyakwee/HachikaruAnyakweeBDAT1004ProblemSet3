Step 5. Discover the Male ratio per occupation and sort it from the most to the least

Step 8. For each occupation present the percentage of women and men


```python
# Step 1: Import the necessary libraries 
import pandas as pd  # Import pandas library and alias it as 'pd' for brevity
import numpy as np   # Import numpy library and alias it as 'np' for brevity
```

```python
# Step 2: Import the dataset from this address.
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

```python
# Step 3: Assign it to a variable called users 
users = pd.read_csv(dataset_url, delimiter='|')  # Using pandas read_csv to load data with '|' delimiter
```

```python
# Step 4: Discover what is the mean age per occupation 
average_age_per_occupation = users.groupby('occupation')['age'].mean()  # Group by occupation and calculate average age
print("Age per Occupation (Mean)")
print(average_age_per_occupation)
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from the most to the least 
# Calculate the male population for each occupation, sorted from highest to lowest
male_census_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count()).sort_values(ascending=False)

# Display the male population per occupation (sorted from highest to lowest)
print("\nMale Population per Occupation (Descending):")
print(male_census_per_occupation)
```

```python
# Step 6: For each occupation, calculate the minimum and maximum ages 
# Compute the age range for each occupation
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Display the minimum and maximum ages per occupation
print("\nAge Range per Occupation:")
print(age_range_per_occupation)
```

```python
# Step 7: Calculate the mean age for each combination of occupation and gender
# Calculate the mean age for each combination of occupation and gender
average_age_by_gender_occupation = users.groupby(['occupation', 'gender'])['age'].mean()

# Display the mean age for each combination of occupation and gender
print("\nOccupation and Gender-Specific Ages (Average)")
print(average_age_by_gender_occupation)
```

```python
# Step 8: Calculate the percentage of women and men for each occupation

# Calculate the percentage of women and men for each occupation
gender_counts = users.groupby(['occupation', 'gender']).size()
occupation_totals = users.groupby('occupation').size()
women_percentage = gender_counts.loc[:, 'F'] / occupation_totals * 100
men_percentage = gender_counts.loc[:, 'M'] / occupation_totals * 100

# Create a DataFrame to display the percentage of women and men per occupation
gender_percentages = pd.DataFrame({'Percentage Women': women_percentage, 'Percentage Men': men_percentage})

# Print the results
print("\nPercentage of women and men per occupation:")
print(gender_percentages)
```

```python
#Step 1. Import the necessary libraries 
# Import the pandas and numpy libraries
import pandas as pd  # Import pandas library and alias it as 'pd' for brevity
import numpy as np   # Import numpy library and alias it as 'np' for brevity
```

```python
#Step 2. Import the dataset from the attached file wind.txt 
# Read the 'wind.txt' file into a DataFrame, considering whitespace as the delimiter and treating 'NaN' as missing values
filepath = r"/Users/hachikaruanyakwee/Library/CloudStorage/OneDrive-Personal/Documents/GEORGIAN COURSEWORK/BDAT 1004/PROBLEM SET 3/wind.txt"
dataset = pd.read_csv(filepath, delim_whitespace=True, na_values='NaN')
```

```python
#Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index. 
# Combine the 'Yr', 'Mo', and 'Dy' columns to create a 'Date' column and convert it to datetime format
dataset['Date'] = pd.to_datetime(dataset[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))

# Set the 'Date' column as the index and drop the 'Yr', 'Mo', 'Dy' columns
dataset = dataset.set_index('Date').drop(['Yr', 'Mo', 'Dy'], axis=1)

# Display the first few rows of the updated dataset
dataset.head()

```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it. 
# Define a function 'fix_year' to adjust years greater than 2000 by subtracting 100 years
def fix_year(x):
    if x.year > 2000:
        return x - pd.DateOffset(years=100)
    return x

# Map the 'fix_year' function to the index of the dataset
dataset.index = dataset.index.map(fix_year)

# Display the first few rows of the updated dataset with adjusted years
dataset.head()

```

```python
#Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns]. 
# Convert the index of the dataset to datetime format using two different methods
dataset.index = pd.to_datetime(dataset.index)
dataset.index = dataset.index.astype('datetime64[ns]')

# Display the updated datetime index
dataset.index

```

```python
#Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below. 
# Calculate the count of null values per location in the 'dataset' DataFrame
null_per_location = dataset.isnull().sum()

# Display the count of missing values per location
null_per_location

```

```python
#Step 7. Compute how many non-missing values there are in total. 
# Calculate the total number of non-missing values in the 'dataset'
total_non_missing = dataset.notnull().sum().sum()

# Display the total number of non-missing values
total_non_missing

```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset. 
# Calculate the average wind speed across all locations and dates in the 'dataset'
avg_windspeed = dataset.mean().mean()

# Display the average wind speed
avg_windspeed


```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days A different set of numbers for each location. 
# Create a DataFrame 'stats_per_location' with statistics (Min, Max, Mean, Std) per location in the 'dataset'
stats_per_location = pd.DataFrame({
    'Min': dataset.min(),
    'Max': dataset.max(),
    'Mean': dataset.mean(),
    'Std': dataset.std()
})

# Display the statistics per location
stats_per_location
```

```python
#Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day. 
# Create a DataFrame 'stats_per_day' with statistics (Min, Max, Mean, Std) per day in the 'dataset'
stats_per_day = pd.DataFrame({
    'Min': dataset.min(axis=1),
    'Max': dataset.max(axis=1),
    'Mean': dataset.mean(axis=1),
    'Std': dataset.std(axis=1)
})

# Display the statistics per day
stats_per_day

```

```python
#Step 11. Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January. 
# Calculate the average windspeed for January, grouped by year
windspeed_avg_january = dataset[dataset.index.month == 1].groupby(dataset[dataset.index.month == 1].index.to_period('Y')).mean()

# Display the average windspeed for January, grouped by year
windspeed_avg_january

```

```python
#Step 12. Downsample the record to a yearly frequency for each location. 
# Resample the 'dataset' to calculate the yearly mean frequency per location
yearly_freq_per_location = dataset.resample('Y').mean()

# Display the yearly mean frequency per location
yearly_freq_per_location
```

```python
#Step 13. Downsample the record to a monthly frequency for each location. 
# Resample the 'dataset' to calculate the monthly mean frequency per location
monthly_freq_per_location = dataset.resample('M').mean()

# Display the monthly mean frequency per location
monthly_freq_per_location
```

```python
#Step 14. Downsample the record to a weekly frequency for each location. 
# Resample the 'dataset' to calculate the weekly mean frequency per location
weekly_freq_per_location = dataset.resample('W').mean()

# Display the weekly mean frequency per location
weekly_freq_per_location
```

```python
#Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks. 
# Create a DataFrame 'weekly_stats_first52' with weekly statistics for the first 52 weeks
weekly_stats_first52 = pd.DataFrame()
weekly_stats_first52['min'] = dataset.resample('W').min().iloc[:52].min(axis=1)
weekly_stats_first52['max'] = dataset.resample('W').max().iloc[:52].max(axis=1)
weekly_stats_first52['mean'] = dataset.resample('W').mean().iloc[:52].mean(axis=1)
weekly_stats_first52['std'] = dataset.resample('W').std().iloc[:52].std(axis=1)

# Display the weekly statistics for the first 52 weeks
weekly_stats_first52
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Specify the file path and read the dataset
filepath = r"/Users/hachikaruanyakwee/Library/CloudStorage/OneDrive-Personal/Documents/GEORGIAN COURSEWORK/BDAT 1004/PROBLEM SET 3/roman-emperor-reigns.csv"
dataset = pd.read_csv(filepath)

# Count the occurrences of 'Assassinated' in the 'Cause_of_Death' column
assassinated_count = dataset['Cause_of_Death'].value_counts()['Assassinated']

# Count the total number of emperors
total_emperors = len(dataset)

# Calculate the percentage of emperors assassinated
percentage_assassinated = (assassinated_count / total_emperors) * 100

# Calculate the percentage of emperors who died from other causes
percentage_other = 100 - percentage_assassinated

# Create a pie chart
labels = ['Assassinated', 'Other Causes']
sizes = [percentage_assassinated, percentage_other]
colors = ['red', 'forestgreen']
explode = (0.1, 0)  # Explode the 'Assassinated' slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Roman Emperors Assassinated')

plt.show()
```

Question 10 
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009. 
 
Don't forget to label your axes! 
Color each dot according to its year. 

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Specify the file path and read the dataset
filepath = r"/Users/hachikaruanyakwee/Library/CloudStorage/OneDrive-Personal/Documents/GEORGIAN COURSEWORK/BDAT 1004/PROBLEM SET 3/arcade-revenue-vs-cs-doctorates.csv"
dataset = pd.read_csv(filepath)

# Extract data from the dataset
revenue = dataset['Total Arcade Revenue (billions)']
doctorates = dataset['Computer Science Doctorates Awarded (US)']
years = dataset['Year']

# Create a scatter plot for the correlation between Arcade Revenue and Computer Science Doctorates
plt.figure(figsize=(10, 6))
for i, year in enumerate(years):
    plt.scatter(revenue[i], doctorates[i], label=year, cmap='viridis', edgecolor='black')

plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Correlation Between Arcade Revenue and Computer Science Doctorates (2000-2009)')
plt.colorbar(label='Year')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
